```python
import pandas as pd
import numpy as np
import os

dfs = {}

path = "../data"

files = os.listdir(path)

for file in files:
    if file.endswith('.csv'):
        df = pd.read_csv(os.path.join(path,file))
        name = os.path.splitext(file)[0]
        dfs[name] = df
```

```python
df_player_stats_scoring_rs = dfs['player_stats_scoring_rs']
df_player_stats_scoring_po = dfs['player_stats_scoring_po']
df_player_stats_advanced_rs = dfs['player_stats_advanced_rs']
df_player_stats_advanced_po = dfs['player_stats_advanced_po']
df_player_stats_misc_rs = dfs['player_stats_misc_rs']
df_player_stats_misc_po = dfs['player_stats_misc_po']
df_player_stats_traditional_po = dfs['player_stats_traditional_po']
df_player_stats_traditional_rs = dfs['player_stats_traditionnal_rs']
df_player_stats_usage_po = dfs['player_stats_usage_po']
df_player_stats_usage_rs = dfs['player_stats_usage_rs']
df_player_stats_defense_po = dfs['player_stats_defense_po']
df_player_stats_defense_rs = dfs['player_stats_defense_rs']


df_player_stats_traditional_po = df_player_stats_traditional_po.drop(columns=['NBA_FANTASY_PTS','WNBA_FANTASY_PTS','NBA_FANTASY_PTS_RANK','WNBA_FANTASY_PTS_RANK'])
df_player_stats_traditional_rs = df_player_stats_traditional_rs.drop(columns=['NBA_FANTASY_PTS','WNBA_FANTASY_PTS','NBA_FANTASY_PTS_RANK','WNBA_FANTASY_PTS_RANK'])

# pd.set_option('display.max_rows', None)
df_player_stats_defense_rs.isnull().sum()
```

```python
# df_player_stats_defense_po[df_player_stats_defense_po["DEF_WS"].isnull()]
df_player_stats_defense_po = df_player_stats_defense_po.dropna(subset=['DEF_WS'])
```

```python
df_player_index = df_player_index.rename(columns={'PERSON_ID': 'PLAYER_ID'})
df_player_index.head()
```

```python
import json
from tqdm import tqdm
```

```python
from pymongo import MongoClient
client = MongoClient("mongodb://root:example@localhost:27017/?retryWrites=true&w=majority")
```

```python
nba_collection = client["dev-nba-stats-app"]["player"]
for index, player in tqdm(df_player_index.iterrows(), desc="Push Player Data to nba collection"):
    player_id = player["PLAYER_ID"]
    player_stats_rs = df_player_stats_traditional_rs[df_player_stats_traditional_rs["PLAYER_ID"] == player_id]
    player_stats_po = df_player_stats_traditional_po[df_player_stats_traditional_po["PLAYER_ID"] == player_id]
    player_stats_advanced_rs = df_player_stats_advanced_rs[df_player_stats_advanced_rs["PLAYER_ID"]== player_id]
    player_stats_advanced_po = df_player_stats_advanced_po[df_player_stats_advanced_po["PLAYER_ID"]== player_id]
    player_stats_scoring_rs = df_player_stats_scoring_rs[df_player_stats_scoring_rs["PLAYER_ID"]== player_id]
    player_stats_scoring_po = df_player_stats_scoring_po[df_player_stats_scoring_po["PLAYER_ID"]== player_id]
    player_stats_misc_rs = df_player_stats_misc_rs[df_player_stats_misc_rs["PLAYER_ID"]== player_id]
    player_stats_misc_po = df_player_stats_misc_po[df_player_stats_misc_po["PLAYER_ID"]== player_id]
    player_stats_usage_rs = df_player_stats_usage_rs[df_player_stats_usage_rs["PLAYER_ID"]== player_id]
    player_stats_usage_po = df_player_stats_usage_po[df_player_stats_usage_po["PLAYER_ID"]== player_id]
    player_stats_defense_rs = df_player_stats_defense_rs[df_player_stats_defense_rs["PLAYER_ID"]== player_id]
    player_stats_defense_po = df_player_stats_defense_po[df_player_stats_defense_po["PLAYER_ID"]== player_id]

    player_data = {
        "PLAYER_ID": player_id,
        "PLAYER_LAST_NAME": player["PLAYER_LAST_NAME"],
        "PLAYER_FIRST_NAME": player["PLAYER_FIRST_NAME"],
        "POSITION": player["POSITION"],
        "HEIGHT": player["HEIGHT"],
        "WEIGHT": player["WEIGHT"],
        "COLLEGE": player["COLLEGE"],
        "COUNTRY": player["COUNTRY"],
        "DRAFT_YEAR": player["DRAFT_YEAR"],
        "DRAFT_ROUND": player["DRAFT_ROUND"],
        "DRAFT_NUMBER": player["DRAFT_NUMBER"],
        "FROM_YEAR": player["FROM_YEAR"],
        "TO_YEAR": player["TO_YEAR"],
        "stats_rs": {
            "traditional":player_stats_rs.to_dict("records"),
            "advanced": player_stats_advanced_rs.to_dict("records"),
            "scoring" : player_stats_scoring_rs.to_dict("records"),
            "misc" : player_stats_misc_rs.to_dict("records"),
            "usage" : player_stats_usage_rs.to_dict("records"),
                    },
        "stats_po": {
            "traditional":player_stats_po.to_dict("records"),
            "advanced": player_stats_advanced_po.to_dict("records"),
            "scoring" : player_stats_scoring_po.to_dict("records"),
            "misc" : player_stats_misc_po.to_dict("records"),
            "usage" : player_stats_usage_po.to_dict("records"),
                    },
    }
    nba_collection.insert_one(player_data)
```
